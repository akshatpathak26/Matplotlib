```python
batsman = pd.read_csv('./sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit sharma Vs Virat kohli Career comparision ')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('Rohit sharma Vs Virat kohli Career comparision ')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='solid',linewidth='2')
plt.title('Rohit sharma Vs Virat kohli Career comparision ')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='solid',linewidth='2',marker='D')
plt.title('Rohit sharma Vs Virat kohli Career comparision ')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='solid',linewidth='2',marker='D',label='RG Sharma')
plt.title('Rohit sharma Vs Virat kohli Career comparision ')

plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

```python
df = pd.read_csv('./batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])- 0.2,df['2015'],width=0.2,color='purple')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+ 0.2,df['2015'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.show()
```
